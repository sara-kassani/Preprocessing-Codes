```python
import numpy as np
from scipy import ndimage, misc
import os, glob
import sys
import skimage
from skimage import feature, data, io
from skimage.morphology import disk
from scipy.misc import imread, imsave
from PIL import Image
```

```python
def get_file_name_dir(rootDir, ext):
 
    # Return file names with certain extension

    return glob.glob1(rootDir, "*." + ext)


def recolor_image(img):
    data = np.array(img)
    r1, g1, b1 = 124,155,5 # Original value
    r2, g2, b2 = 0, 0, 0 # Value that we want to replace it with

    red, green, blue = data[:,:,0], data[:,:,1], data[:,:,2]
    mask = (red == r1) & (green == g1) & (blue == b1)
    
    data[:,:,:3][mask] = [r2, g2, b2]
    im = Image.fromarray(data)
    recolored = im
    return recolored
```

```python
"""
Get name images to preprocessing
--------------------------------
"""
input_path = 'data/images/'
input_extension = 'png'

inputs_files = sorted(get_file_name_dir(input_path, input_extension))

print(('**')*30)
print('********************* Preprocessing *************************')
print(('**')*30)

"""
Iterate over the list of images
-------------------------------
"""
cont = 1

for file_name in inputs_files:
    print('Preprocessing: ' + file_name + ': ' + str(cont))
    cont += 1
    # Get the image
    image = Image.open(input_path + file_name)
    # Crop Image
    image = recolor_image(image)
#     image = Image.fromarray(image)
     # Save Image
    imsave('data/processed' + file_name, image)
```
